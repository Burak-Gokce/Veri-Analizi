```python
import pandas as pd
import numpy as np
import seaborn as sns
```

## Veri setini yükle ve okut.

```python
df = sns.load_dataset("diamonds")
```

## Apply fonksiyonunu kullanarak örnekler yap.

```python
buyuk_harf = lambda x: x.upper()
```

```python
df["cut"].apply(buyuk_harf)
```

```python
kucuk_harf = lambda x: x.lower()
```

```python
df["color"].apply(kucuk_harf)
```

```python
ortalama = lambda x: x.mean()
```

```python
df[["depth","table"]].apply(ortalama)
```

```python
isfound = lambda x: x.startswith('I')
```

```python
df[df["cut"].apply(isfound)]
```

```python
sayim = lambda x: x.count('E')
```

```python
df["color"].apply(sayim).sum()
```

```python
df.groupby("color").get_group("E")
```

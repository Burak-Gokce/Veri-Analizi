## Gruplandırma Yöntemleri

```python
df.head()
```

```python
df.groupby("sex").get_group("female")[df["alive"]=="yes"]
#cinsiyeti kadın olup hayatta kalanları çağırdık.
```

```python
df.groupby("sex").get_group("female")[df["alive"]=="no"]
#cinsiyeti kadın olup hayatta olmayanları çağırdık.
```

```python
df.groupby("class").get_group("First")
```

```python
df.groupby("class").get_group("Third")
```

```python
df.groupby("class").sum()
```

```python
df.groupby("class")[["survived"]].mean()
#yolcuların sınıflara göre hayatta kalma ortalamalarını elde ettik.
```

```python
df.groupby("pclass").get_group(3)[df["alive"]=="no"]
```

```python
df.groupby("class")[["survived"]].sum()
```

```python
c = df.groupby("pclass")
c.sum()
```

```python
df.groupby("class")["age"].mean()
#sınıflara göre yaş ortalaması aldık.
```

```python
df.groupby(["class","sex"])[["age"]].mean()
#sınıfları ve cinsiyetleri beraber sınıflandırdıktan sonra yaş ortalamalarını aldık.
```

```python
df.groupby("class")["age"].max()
```

```python
df.groupby(["class","sex"])[["age"]].max()
```

```python
df["sex"].value_counts()
#her bir değer kaç kere tekrar etmiş onu veriyor bize.
```

```python
df["alive"].value_counts()
```

```python
df.groupby(["class","sex"])[["age"]].agg([np.sum,np.mean,np.max,np.min])
#aggregate fonksiyonuyla toplam,minimum,maksimum ve ortalama değerlerini aldık.
```

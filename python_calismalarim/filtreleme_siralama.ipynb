## Veri setini indir ve atama yap.

```python
pd.read_csv("readingdata/nba.txt")
```

```python
df = pd.read_csv("readingdata/nba.txt")
```

## Maaşı 1000000 dolar üstü olan oyuncu sayısı?

```python
df[df["Salary"] >= 1000000]
```

```python
df[df["Salary"] >= 1000000].shape[0]
```

## Maaşları sıralayınız.

```python
df[df["Salary"] >= 1000000].sort_values("Salary")
```

## Maaşları büyükten küçüğe doğru sıraladık.

```python
df[df["Salary"] >= 1000000].sort_values("Salary",ascending=False)
```

## En yüksek maaşı alan oyuncu kimdir?

```python
df[df["Salary"] == df["Salary"].max()]
```

## En düşük maaşı alan oyuncu kimdir?

```python
df[df["Salary"] == df["Salary"].min()]
```

## En yaşlı oyuncu kimdir?

```python
df[df["Age"] == df["Age"].max()]["Name"]
```

## En çok sayı yapan oyuncu kimdir?

```python
df[df["Number"] == df["Number"].max()]
```

## En az sayı yapan oyuncu kimdir?

```python
df[df["Number"] == df["Number"].min()]
```
